```python
import numpy as np
import matplotlib.pyplot as plt

# Generar la serie x
x = np.arange(1, 101)

# Generar la serie y con un pequeño ruido aleatorio
y = np.linspace(1, 1000, 100) + np.random.normal(0, 20, 100)

# Graficar los datos
plt.scatter(x, y, label='Datos sintéticos', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Serie sintética')
plt.legend()
plt.show()

# Mostrar las primeras 10 observaciones
list(zip(x, y[:10]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([14.275, 30.171, 8.005, 33.310, 6.226])

tree = DecisionTreeRegressor(max_depth=3)  # Árbol con 2 niveles
tree.fit(x, y)

plt.scatter(x, y, color='blue', label="Datos originales")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Datos Originales")
plt.legend()
plt.show()

x_range = np.linspace(0.5, 5.5, 100).reshape(-1, 1)  # Valores de prueba
y_pred = tree.predict(x_range)  # Predicción del árbol

plt.scatter(x, y, color='blue', label="Datos originales")
plt.plot(x_range, y_pred, color='red', linewidth=2, label="Modelo del Árbol")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Árbol de Decisión para Regresión")
plt.legend()
plt.show()

```

```python
plt.scatter(x, y, color='blue', label="Datos originales")

# Obtener los puntos de corte del árbol
thresholds = tree.tree_.threshold[tree.tree_.threshold > 0]  # Filtrar valores válidos

for threshold in thresholds:
    plt.axvline(x=threshold, color='green', linestyle="--", label=f"Corte en x={threshold:.2f}")

plt.plot(x_range, y_pred, color='red', linewidth=2, label="Modelo del Árbol")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Puntos de Corte del Árbol")
plt.legend()
plt.show()
```
